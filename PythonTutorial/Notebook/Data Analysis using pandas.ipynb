```python
### Arithmetic Operators with new coloumn enter as Total_Marks at last
data['Total_Marks'] = data['Paper_1'] + data['Paper_2'] + data['Paper_3'] + data['Paper_4']
```

```python
# Adding Percentages coloumn 
data['Percentage'] = data['Total_Marks']/6
```

```python
#making final result coloumn using def fun
def result_status(marks):
    if marks>=33:
        return 'Pass'
    else:
        return 'Fail'
```

```python
data['Paper_1_Status'] = data.Paper_1.apply(result_status)
```

```python
# Using Lambda function
data['Lambda_Marks'] = data.apply(lambda x: x['Paper_1'] + x['Paper_2'] + x['Paper_3'] + x['Paper_4'] + x['Paper_6'])
```
